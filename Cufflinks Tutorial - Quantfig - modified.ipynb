```python
import cufflinks as cf
```

```python
df=cf.datagen.ohlcv()
```

```python
qf=cf.QuantFig(df,title='First Quant Figure',legend='top',name='GS')
qf.add_bollinger_bands()
qf.iplot()
```

```python
qf.add_sma([10,20],width=2,color=['green','lightgreen'],legendgroup=True)
qf.add_rsi(periods=20,color='java')
qf.add_bollinger_bands(periods=20,boll_std=2,colors=['magenta','grey'],fill=True)
qf.add_volume()
qf.add_macd()
qf.iplot()
```

# volume の bar chart のホバーを変える

```python
qf=cf.QuantFig(df,title='First Quant Figure',legend='top',name='GS')
qf.add_volume(hovertext=[f"A:\n{i}"for i in range(len(df))])
qf.add_macd()
qf.iplot()
```

複数行にわたる場合は '\<br\>' 区切り。

```python
qf=cf.QuantFig(df,title='First Quant Figure',legend='top',name='GS')
qf.add_volume(hovertext=[f"A:<BR>{i}"for i in range(len(df))])
qf.add_macd()
qf.iplot()
```

```python
qf=cf.QuantFig(df,title='First Quant Figure',legend='top',name='GS')
qf.add_volume(hovertext=[f"A:<BR>{i}<BR>{i}<BR>{i}<BR>{i}"for i in range(len(df))])
qf.add_macd()
qf.iplot()
```

```python
text = [
    f"A:<BR>{row['open']}<BR>{row['high']}<BR>{row['low']}<BR>{row['close']}" for i, row in df.iterrows() 
]
text
```

```python
# candlestick
df['high'] = df[['open', 'close']].max(axis=1)
df['low'] = df[['open', 'close']].min(axis=1)
```

```python
text = [
    f"A:<BR>{int(row['open'])}<BR>{int(row['close'])}" for i, row in df.iterrows() 
]
text
```

```python
qf=cf.QuantFig(df,title='First Quant Figure',legend='top',name='GS')
qf.add_volume(hovertext=text)
qf.iplot()
```
